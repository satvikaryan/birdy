```python
# y, sr = librosa.load("/home/satvikaryan/design_assignment/wav/dash_1.wav")
audio_data1 = "/home/satvikaryan/birdy/wav/dash_1.wav"
audio_data2 = "/home/satvikaryan/birdy/wav/dash_2.wav"
ipd.Audio(audio_data1)
```

```python
audio1, sample_rate1 = librosa.load(audio_data1)
```

```python
audio2, sample_rate2 = librosa.load(audio_data2)
```

```python
frameSize = 2048
hopSize = 512
```

```python
stft_audio1 = librosa.stft(audio1, n_fft = frameSize, hop_length = hopSize)
```

```python
stft_audio2 = librosa.stft(audio2, n_fft = frameSize, hop_length = hopSize)
```

```python
stft_audio1.shape
```

```python
stft_audio2.shape
```

```python
y_audio1 = np.abs(stft_audio1) ** 2
y_audio1.shape
```

```python
y_audio2 = np.abs(stft_audio2) ** 2
y_audio2.shape
```

```python
def plot_spectrogram(y, sr, hop_length, y_axis = "linear"):
    plt.figure(figsize = (25,10))
    librosa.display.specshow(y, sr = sr, hop_length = hop_length, x_axis = "time", y_axis = y_axis)
    plt.colorbar(format="%+2.f")
```

```python
plot_spectrogram(y_audio1, sample_rate1, hopSize)
```

```python
plot_spectrogram(y_audio2, sample_rate2, hopSize)
```

```python
y_log_audio1 = librosa.power_to_db(y_audio1)
plot_spectrogram(y_log_audio1, sample_rate1, hopSize)
```

```python
y_log_audio2 = librosa.power_to_db(y_audio2)
plot_spectrogram(y_log_audio2, sample_rate2, hopSize)
```

```python
plot_spectrogram(y_log_audio1, sample_rate1, hopSize, y_axis = "log")
```

```python
plot_spectrogram(y_log_audio2, sample_rate2, hopSize, y_axis = "log")
```
